```python
import pandas as pd
import numpy as np
import tushare as ts
from pyecharts.charts import Map
from pyecharts import options as opts
```

```python
df1 = pd.read_csv(r'C:\Users\Administrator\Desktop\Task_4_lianjia_sale.csv', encoding='utf-8')
data1 = df1.groupby('区域')['价格'].agg([('区域平均价格','mean')])
df_avg = data1.sort_values(by='区域平均价格')
districts = ['金山区', '奉贤区', '嘉定区', '松江区', '青浦区', '宝山区', '闵行区 ', '浦东新区', '普陀区', '杨浦区', '虹口区', '长宁区', '静安区', '徐汇区', '黄浦区']
avg_price = [round(i,2) for i in df_avg["区域平均价格"].values.tolist()]
```

```python
map_data1 = [list(z) for z in zip(districts, avg_price)]
map1 = Map()
map1.set_global_opts(
    title_opts=opts.TitleOpts(title="上海各区二手房均价"),
    visualmap_opts=opts.VisualMapOpts(max_=max(avg_price), min_=min(avg_price))
)
map1.add("", map_data1, "上海")
map1.render_notebook()
```

```python
df2 = ts.get_area_classified()
data2 = df2.groupby("area")["name"].agg([("上市公司数","count")])
provinces = data2.index.values.tolist()
comp_count = data2["上市公司数"].values.tolist()
map_data2=[list(z) for z in zip(provinces, comp_count)]
map2 = Map()
map2.set_global_opts(
    title_opts=opts.TitleOpts(title="中国个省上市公司分布情况（除港澳台）"),
    visualmap_opts=opts.VisualMapOpts(max_=max(comp_count),min_=min(comp_count), is_piecewise=True)
)
map2.add("", map_data2, maptype="china")
map2.render_notebook()
```
