```python
import pandas as pd
import numpy as np
from pyecharts import options as opts
from pyecharts.charts import Bar, Pie, Page, Funnel
from pyecharts.globals import ThemeType
```

```python
df1 = pd.read_csv(r'C:\Users\Administrator\Desktop\Task_4_lianjia_sale.csv', encoding='utf-8')
df1.head(3)
```

```python
data1 = df1.groupby('区域')['价格'].agg([('区域平均价格','mean')])
df_avg = data1.sort_values(by='区域平均价格')
df_avg.head()
```

```python
x1 = df_avg.index.tolist()
y1 = [round(i,2) for i in df_avg["区域平均价格"].values.tolist()]
bar1 = Bar().add_xaxis(x1).add_yaxis("",y1)
bar1.set_global_opts(title_opts=opts.TitleOpts(title="上海各区域平均价格"))
bar1.reversal_axis()
bar1.render_notebook()
```

```python
df_zhuangxiu = df1.groupby('装修类型')['房型'].agg([('房源数量','count')])
df_zhuangxiu.head()
```

```python
x2 = df_zhuangxiu.index.tolist()
y2 = df_zhuangxiu['房源数量'].values.tolist()
data = [list(z) for z in zip(x2, y2)]
pie = Pie().add("", data)
pie.set_global_opts(title_opts=opts.TitleOpts(title="装修类型占比情况"))
pie.set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {d}%"))
pie.render_notebook()
```
